## **FUNCIÓN 1: PEDIR NOMBRE VÁLIDO**

```python
def pedir_nombre_valido():
    """
    Pide el nombre completo del jugador.
    Permite mayúsculas en mitad del nombre: "McCollum".

    Permite apóstrofes: "Shaquille O'Neal".
    Permite puntos '.': "O.J. Mayo".
    Permite guiones: '-': "Shai Gilgeous-Alexander".
    """

    while True:
        nombre = input("Introduce el nombre del jugador: ").strip()
        # Strip -> Elimina posibles espacios al inicio y al final.

        # En un abordaje inicial también usé .title, para evitar que el usuario introduzca todo en mayúsculas.
        # Pero eso provocaba nombre en fallos como "LeBron" -> "Lebron" o "VanVleet" -> "Vanvleet".
        # Respetamos lo que introduzca el usuario.

        nombre = " ".join(nombre.split()) # Elimina espacios de sobra entre nombre y apellido(s).

        # 1. NOMBRE NO VACÍO.
        if nombre == "":
            print("ERROR. El nombre no puede estar vacío. Prueba de nuevo.\n")
            continue

        # 2. NO EMPEZAR NI ACABAR CON SÍMBOLOS.
        if nombre[0] in "-'." or nombre[-1] in "-'.":
            print("ERROR. El nombre no puede empezar ni acabar con símbolos. Prueba de nuevo.\n")
            continue

        # 3. CARACTERES PERMITIDOS -> NO SE PERMITEN NÚMEROS NI SÍMBOLOS RAROS.
        # Sí se permiten cualquier tipo de letra con acento: "Dončić", "Jokić".

        if any(not (c.isalpha() or c in " .'-") for c in nombre):
            print("ERROR. El nombre contiene caracteres no válidos. Prueba de nuevo.\n")
            continue

        # 4. SECUENCIAS INVÁLIDAS -> No puede haber dos guiones o dos puntos consecutivos.
        invalidos = ["--", "''", "..", "- ", " -", "' ", " '", "-'", "'-", "-.", ".-", "'.", ".'"]
        if any(x in nombre for x in invalidos):
            print("ERROR. Uso incorrecto de guiones, apóstrofes o puntos. Prueba de nuevo.\n")

            continue

        return nombre
```

```python
# PRUEBA
nombre_prueba = pedir_nombre_valido()
print(f"Tu jugador se llama: {nombre_prueba}.")
```

## **FUNCIÓN 2: PEDIR TIPO DE TIRO**

```python
def pedir_tipo_tiro():
    while True:
        try:
            tipo_de_tiro = int(input("Introduce el tipo de tiro (1/2/3): "))

            if tipo_de_tiro not in [1, 2, 3]:
                print("ERROR. El tipo de tiro debe ser 1, 2 o 3.\n")
                continue

            return tipo_de_tiro  # Solo se devuelve cuando es válido

        except ValueError:
            print("ERROR. Debes introducir un número entero válido.\n")
```

## **CALCULADORA 0: PORCENTAJE DE TIRO DE UN JUGADOR EN UN PARTIDO**
## **CALCULADORA FINAL**

```python
def calcular_porcentaje_tiro():

    print("*NOMBRE*\n")
    print("En primer lugar, introduce el nombre de tu jugador.")
    print("Este solo debe contener letras, espacios y guiones. Por favor, no introduzcas caracteres numéricos")
    print("Si introduces un formato incorrecto, te solicitaré el nombre hasta que introduzcas uno válido.\n")

    nombre_jugador = pedir_nombre_valido()

    print("\n------------------------------------------------------------------------------------------------------------\n")

    print("*TIPO DE TIRO*")
    print("\n\nEn segundo lugar, introduce el tipo de tiro. Para ello, responde con un único dígito: '1', '2' o '3'.\n")
    print("Si introduces '1', trabajaremos con tiros libres.")
    print("Si introduces '2', trabajaremos con tiros de dos puntos.")
    print("Si introduces '3', trabajaremos con tiros de tres puntos.\n")

    tipo_tiro = pedir_tipo_tiro()


    seguir = "S"

    while seguir == "S":
      tiros_anotados = int(input(f"\nIntroduce los tiros de {tipo_tiro}p anotados: "))
      tiros_intentados = int(input(f"Introduce los tiros de {tipo_tiro}p intentados: "))

      if ((tiros_intentados<0) or (tiros_anotados<0)):
        print("ERROR. No puedes introducir un número negativo de lanzamientos. Prueba otra vez-\n" )
        continue

      elif (tiros_anotados > tiros_intentados):
        print("ERROR. No puedes anotar más tiros de los que intentas. Prueba otra vez.\n" )
        continue

      elif (tiros_intentados == 0):
        print("\nNo hubo tiros intentados. No se puede calcular porcentaje. Por convenio, esto se representa como:")
        print(f"Porcentaje de tiro de {tipo_tiro}p: 0%\n")

      else:
        porcentaje_tiro = (tiros_anotados/tiros_intentados)*100

        print("\n------------------------------------------------------------------------------------------------------------\n")
        print("*JUGADOR & PORCENTAJE DE TIRO*\n")
        print(f"\nJugador: {nombre_jugador}")
        print(f"Porcentaje de tiro de {tipo_tiro}p: {porcentaje_tiro:.2f}%\n")

      seguir = input(f"¿Quieres seguir introduciendo tiros de {tipo_tiro}p?\nResponde 'S' para seguir.\nResponde otra tecla cualquiera para terminar: ").upper()

      # Devolvemos el último porcentaje calculado

    return {
         "NOMBRE": nombre_jugador,
         "TIPO_TIRO": tipo_tiro,
         "PORCENTAJE": porcentaje_tiro
      }
```

## **FUNCIÓN 3.3: PEDIR DORSAL DEL JUGADOR**

```python
def pedir_dorsal_valido():
    """
    Pide un dorsal válido según reglamento ACB:
    Permite '00' y del '0' al '99'.
    Devuelve el dorsal como STRING, NO como ENTERO.
    """

    print("Introduce el dorsal del jugador. Valores permitidos: 00, 0, 1, ..., 99.\n---\n")

    while True:
        dorsal = input("Dorsal: ").strip()

        # Solo dígitos
        if not dorsal.isdigit():
            print("ERROR. Caracter no válido. El dorsal debe contener solo números.\n---\n")
            continue

        # Longitud máxima 2
        if len(dorsal) > 2:
            print("ERROR. El dorsal debe ser un número de máximo dos cifras.\n---\n")
            continue

        # Rango válido
        valor = int(dorsal)
        if valor < 0 or valor > 99:
            print("ERROR. El dorsal debe estar entre 0 y 99.\n---\n")
            continue

        # Caso especial permitido: '0' y '00'
        if dorsal in ("0", "00"):
            return dorsal

        # Evitar ceros a la izquierda tipo '01'
        if dorsal.startswith("0"):
            print("ERROR. Puedes usar '0' o '00', pero el resto de dorsales no pueden tener ceros a la izquierda. Ej: '01' no está permitido.\n---\n")
            continue

        return dorsal
```

```python
# PRUEBA
dorsal_prueba=pedir_dorsal_valido()
print(f"El dorsal de tu jugador es: {dorsal_prueba}")
```

## **FUNCIÓN 4.1: PEDIR ESTADÍSTICA COMPLETA ACB**

```python
def pedir_estadisticas_jugador():
    """
    Pide las estadísticas completas de un jugador en un partido. Modelo; Boxscore ACB.
    ESTRUCTURA DE LA FUNCIÓN:

    1. IDENTIFICACIÓN DEL JUGADOR. -> PEDIR NOMBRE Y DORSAL.
    Al programarlo inicialmente, pedía nombre y dorsal dentro del bucle. Podría hacerse así, no hay problema.
    PERO, de cara a validar datos, me he dado cuenta de que es mejor hacerlo fuera. -> Lo dejo comentado.

    2. BUCLE WHILE.
      2.1. Solicitamos estadísticas.
      2.2. Realizamos validaciones.

    3. RESULTADO FUNCIÓN -> DEVOLVER ESTADÍSTICAS COMPLETAS EN UN DICCIONARIO.
    """

    # 1. IDENTIDAD.
    # nombre = pedir_nombre_valido()
    # dorsal = pedir_dorsal_valido()

    # 2.
    while True:
        try:

            # -------------------------------------------------
            # 2.1 PEDIR ESTADÍSTICAS.
            # -------------------------------------------------

            # 2.1 - TIEMPO.
            minutos = int(input("\nMinutos jugados: "))
            segundos = int(input("Segundos jugados: "))

            # 2.1 - TIROS (anotados -> intentados).
            tl_anot = int(input("TL anotados: "))
            tl_int = int(input("TL intentados: "))
            t2_anot = int(input("2P anotados: "))
            t2_int = int(input("2P intentados: "))
            t3_anot = int(input("3P anotados: "))
            t3_int = int(input("3P intentados: "))

            # 2.1 - REBOTES.
            reb_o = int(input("Rebotes ofensivos: "))
            reb_d = int(input("Rebotes defensivos: "))

            # 2.1 - ASISTENCIAS, PERIDAS, RECUPERACIONES (i.e. robos).
            ast = int(input("Asistencias: "))
            perd = int(input("Pérdidas: "))
            rec = int(input("Recuperaciones: "))

            # 2.1 - TAPONES.
            tap_f = int(input("Tapones a favor: "))
            tap_c = int(input("Tapones recibidos: "))

            # 2.1 - MATES.
            mates = int(input("Mates realizados: "))

            # 2.1 - FALTAS.
            fc = int(input("Faltas cometidas: "))
            fr = int(input("Faltas recibidas: "))

            # 2.1 - IMPACTO (+-).
            plus_minus = int(input("Plus/Minus: "))

            # CÁLCULO DE TOTALES: Minutos, Puntos & Rebotes.
            min_total = minutos + segundos / 60
            puntos = tl_anot * 1 + t2_anot * 2 + t3_anot * 3
            reb_tot = reb_o + reb_d


            # CÁLCULO DE TOTALES: Valoración ACB
            tiros_fallados = (tl_int - tl_anot) + (t2_int - t2_anot) + (t3_int - t3_anot)
            valoracion = puntos + reb_tot + ast + rec + tap_f + fr - ( tiros_fallados + perd + tap_c + fc)

            # -------------------------------------------------
            # 2.2 VALIDACIONES.
            # -------------------------------------------------

            # 2.2 - NO VALORES NEGATIVOS.
            if min(minutos, segundos, tl_anot, tl_int, t2_anot, t2_int,
                   t3_anot, t3_int, reb_o, reb_d, ast, perd, rec,
                   tap_f, tap_c, mates, fc, fr) < 0:
                print("ERROR. No se permiten valores negativos.")
                continue

            # 2.2 - SEGUNDOS EN RANGO.
            if segundos >= 60:
                print("ERROR. Los segundos deben estar entre 0 y 59.")
                continue

            # 2.2 - TIROS COHERENTES.
            if tl_anot > tl_int or t2_anot > t2_int or t3_anot > t3_int:
                print("ERROR. Los tiros anotados no pueden superar a los intentados.")
                continue

            # 2.2 - MÁXIMO FALTAS PERSONALES.
            if fc > 5:
                print("ERROR. Un jugador no puede cometer más de 5 faltas.")
                continue

            # 2.2 - TAPONES RECIBIDOS COHERENTES CON TIROS DE CAMPO INTENTADOS.
            tiros_campo_int = t2_int + t3_int
            if tap_c > tiros_campo_int:
                print("ERROR. No se pueden recibir más tapones que tiros intentados.")
                continue

            # 2.2 - MINUTOS TOTALES COHERENTES.
            if min_total > 90:
                print("ERROR. Minutos jugados no realistas para un partido ACB.")
                continue

            break

        # 2.2 - DATOS DE TIPO 'int'.
        except ValueError:
            print("ERROR. Introduce valores numéricos válidos.\n")

    #3. RESULTADO FUNCIÓN -> ESTADÍSTICAS EN DICCIONARIO.

    return {
        "PTS": puntos,
        "TL_anot": tl_anot, "TL_int": tl_int,
        "2P_anot": t2_anot, "2P_int": t2_int,
        "3P_anot": t3_anot, "3P_int": t3_int,
        "REB_O": reb_o, "REB_D": reb_d,
        "REB_TOT": reb_tot,
        "AST": ast,
        "PERD": perd,
        "REC": rec,
        "TAP_F": tap_f, "TAP_C": tap_c,
        "MAT":mates,
        "FC": fc, "FR": fr,
        "PLUS_MINUS": plus_minus,
        "VAL": valoracion,
        "MIN": min_total,
        "MIN_MMSS": f"{minutos}' {segundos}''"
    }
```

## **FUNCIÓN 4.2: AÑADIR NOMBRE Y DORSAL A ESTADÍSTICAS**

```python
def añadir_nombre_y_dorsal_stats(nombre,dorsal,stats):

    stats_completas = {
        "NOMBRE": nombre,
        "DORSAL": dorsal,
        **stats
    }

    return stats_completas
```

## **FUNCIÓN 5: CALCULAR MÉTRICAS INDIVIDUALES -> ESTADÍSTICA CONVENCIONAL**

```python
# PUNTOS TOTALES -> Sumar cada tipo de tiro anotado, multiplicando por su valor.
def puntos(stats):
    return (stats["TL_anot"] + 2*stats["2P_anot"] + 3*stats["3P_anot"])

# REBOTES TOTALES -> Sumar rebotes ofensivos y defensivos.
def rebotes(stats):
    return (stats["REB_O"] + stats["REB_D"])

# TIROS FALLADOS -> Restar los tiros anotados a los tiros intentados
def tiros_fallados(stats):
    return (stats["TL_int"] - stats["TL_anot"]) + (stats["2P_int"] - stats["2P_anot"]) + (stats["3P_int"] - stats["3P_anot"])

# PORCENTAJES TIRO
def porcentajes_tiro(stats):
    pct_tl = (stats["TL_anot"]/stats["TL_int"]*100) if stats["TL_int"]>0 else 0
    pct_2p = (stats["2P_anot"]/stats["2P_int"]*100) if stats["2P_int"]>0 else 0
    pct_3p = (stats["3P_anot"]/stats["3P_int"]*100) if stats["3P_int"]>0 else 0
    return pct_tl, pct_2p, pct_3p

# VALORACIÓN ACB -> Estadísticas positivas suman y negativas restan. Es decir:
# VAL = + [ pts + reb + asist + recup + tapones_favor + faltas_recibidas ]                (Stats que suman)
#       - [tiros fallados + pérdidas + tapones_contra + faltas_cometidas ]                (Stats que restan)

def valoracion_acb(stats):
    return (puntos(stats) + rebotes(stats) + stats["AST"] + stats["REC"] + stats["TAP_F"] + stats["FR"] - tiros_fallados(stats) - stats["PERD"] - stats["TAP_C"] - stats["FC"])
```

```python
# POSESIONES -> POS = (TIROS CAMPO INTENTADOS) + 0,44*TIROS LIBRES INTENTADOS + PERDIDAS - REBOTES OFENSIVOS
def posesiones(stats):
    return (stats["2P_int"] + stats["3P_int"] + 0,44*stats["TL_int"] + stats["PERD"] - stats["REB_O"])
```

## **FUNCIÓN 6: VALIDAR IDENTIDAD JUGADORES**

```python
def validar_identidad(nombre, dorsal, equipo):
    """
    FUNCIÓN DE VALIDACIÓN -> Devuelve (TRUE/FALSE) + MENSAJE.
    Si devuelve TRUE, el mensaje es vacío ("").
    Si devuelve FALSE, el mensaje es MENAJE DE ERROR.
    """
    # El nombre es la clave que usamos en el diccionario general "equipo" para identificar a cada jugador.

    # 1. EVITAR NOMBRE DUPLICADO.
    if nombre in equipo:
        return False, "ERROR. Ya existe un jugador con ese nombre."

    # 2. EVITAR DORSAL DUPLICADO.
    for jugador in equipo.values():
        if jugador["DORSAL"] == dorsal:
            return False, f"ERROR. El dorsal {dorsal} ya está asignado a un jugador."

    return True, ""
```

## **FUNCIÓN 7.1: PEDIR TEXTOS (NOMBRES)**


```python
import re

# 're' es el módulo de expresiones regulares en Python.
# Estas expresiones son patrones que permiten buscar o validar cadenas de texto de manera muy flexible.


def pedir_texto_valido(mensaje):
    """
    Pide al usuario un texto válido para nombre de equipo, competición, etc.
    Se permite: letras, números, acentos, espacios, guiones, 'ª', 'º', '/', '(', ')', '.', ','.
    No se permite: texto vacío, guiones al inicio/final ni dobles guiones.
    """
    while True:
        texto = input(mensaje).strip()  # Quitamos espacios al inicio y final.

        # TEXTO NO VACÍO.
        if texto == "":
            print("ERROR. El nombre no puede estar vacío. Prueba de nuevo.\n")
            continue

        # GUIONES INCORRECTOS.
        if texto.startswith("-") or texto.endswith("-") or "--" in texto or "- " in texto or " -" in texto:
            print("ERROR. Has introducido un guión no válido. Prueba de nuevo.\n")
            continue

        # COMPROBACIÓN CARACTÉRES VÁLIDOS.
        # re.fullmatch(patron, texto) -> Comprueba que todo el 'texto' coincide exactamente con el 'formato' (i.e. caracteres válidos)
        # re.fullmatch(patron, texto) -> Devuelve un 'Match object' si se cumple, o 'None'  si no se cumple.

        # PATRÓN COMPLETO.
        patron = r"[A-Za-z0-9ÁÉÍÓÚÜáéíóúüñÑªº \-()/.,]+"

        # VALIDACIÓN -> con re.fullmatch
        if not re.fullmatch(patron, texto):
            print("ERROR. Solo se permiten letras, números, espacios, guiones, paréntesis, '/', '.', ',' y 'ª/º'. Prueba de nuevo.\n")
            continue

        #Si NO se cumple el formato pedido -> Texto no válido -> Repetimos el bucle (saltamos a siguiente iteración con continue).
        #Si SÍ se cumple el formato pedido -> Texto válido -> El bucle acaba. La función devuelve el texto introducido.

        return texto
```

## **FUNCIÓN 7.2: PEDIR TEMPORADA**



```python
def pedir_temporada_valida():

    """
    Pide al usuario una temporada.

    FORMATOS VÁLIDOS:    ->                             EJEMPLOS VÁLIDOS:
      - "aaaa"           ->                             Ej: "2025".
      - "aaaa-aaaa"      -> Solo años consecutivos.     Ej: "2025-2026".

    FORMATOS NO VÁLIDOS: ->                             EJEMPLOS NO VÁLIDOS:
      - "aaaa/aaaa"      ->                             Ej: "2025/2026".
      - Año inicio > Año fin  ->                        Ej: "2025-2024".
      - Años no consecutivos  ->                        Ej: "2020-2023".

    DEBE CUMPLIRSE -> AÑO INICIO + 1 = AÑO FIN

    """

    while True:
        temporada = input("Introduce la temporada (aaaa o aaaa-aaaa): ").strip()

        # PATRÓN para: 4 dígitos o (4 dígitos - 4 dígitos).
        patron = r"\d{4}(-\d{4})?"
        if not re.fullmatch(patron, temporada):
            print("ERROR. Formato incorrecto. Debe ser 'aaaa' o 'aaaa-aaaa'. Ej: 2025 o 2020-2021.\n")
            continue

        # VALIDACIÓN DE RANGO: Si el formato es (año inicio-año fin), los años deben ser consecutivos, con "año fin = año inicio +1".
        if "-" in temporada:
            inicio, fin = map(int, temporada.split("-"))
            if fin != inicio + 1:
                print("EROR. En un rango de temporada, el segundo año debe ser consecutivo al primero. Ej: 2020-2021.\n")
                continue

        return temporada
```

## **FUNCIÓN 7.3: PEDIR FECHA**

```python
import datetime

def pedir_fecha_valida(mensaje="Introduce la fecha (dd/mm/aaaa): "):
    """
    Pide una fecha en formato dd/mm/aaaa y valida que sea una fecha real.
    """
    while True:
        fecha_str = input(mensaje).strip()
        try:
            fecha = datetime.datetime.strptime(fecha_str, "%d/%m/%Y")
            return fecha   # SIEMPRE devuelve datetime
        except ValueError:
            print("ERROR. Formato inválido o fecha no real. Usa dd/mm/aaaa. Prueba de nuevo.\n")

```

```python
def pedir_estadisticas_equipo():
    """

    Función principal para introducir estadísticas de un partido.

    Devuelve un diccionario con:
    - Datos del partido: nombre equipo, rival, competición, fecha, temporada.
    - Jugadores con sus estadísticas.
    - Fila de estadísticas de equipo (rebotes, pérdidas, faltas técnicas, etc) ->
    -> Sirve para contabilizar rebotes del equipo que no se anotan a ningún jugador, técnicas al banquillo, etc

    """
    equipo = {}  # Diccionario de jugadores

    # 1. DATOS DEL PARTIDO
    print("\n--- DATOS DEL PARTIDO ---\n")
    nombre_equipo = pedir_texto_valido("Introduce el nombre de tu equipo: ")
    nombre_rival = pedir_texto_valido("Introduce el nombre del equipo rival: ")
    competicion = pedir_texto_valido("Introduce el nombre de la competición: ")
    temporada = pedir_temporada_valida()
    fecha = pedir_fecha_valida()  # Devuelve objeto datetime

    # 2. NÚMERO DE JUGADORES
    numero_jugadores = pedir_numero_jugadores()

    print(f"\nPodrás introducir {numero_jugadores} jugadores. Luego se ordenarán por dorsal ascendente.\n")

    # 3. INTRODUCIR JUGADORES
    for i in range(numero_jugadores):
        print(f"\n--- JUGADOR {i+1} ---\n")

        # 3.1 PEDIR NOMBRE Y DORSAL
        while True:
            nombre = pedir_nombre_valido()
            dorsal = pedir_dorsal_valido()
            valido, mensaje = validar_identidad(nombre, dorsal, equipo)
            if not valido:
                print(mensaje)
                continue
            break

        # 3.2 PEDIR ESTADÍSTICAS INDIVIDUALES
        stats = pedir_estadisticas_jugador()

        # 3.3 AÑADIR NOMBRE Y DORSAL a estadísticas
        stats_completas = añadir_nombre_y_dorsal_stats(nombre, dorsal, stats)

        # 3.4 GUARDAR EN EL DICCIONARIO DEL EQUIPO
        equipo[nombre] = stats_completas

    # 4. FILA DE ESTADÍSTICAS DEL EQUIPO (sin asignar a jugadores)
    print("\n--- ESTADÍSTICAS DE EQUIPO (no asignadas a jugadores) ---\n")
    stats_equipo = pedir_estadisticas_equipo_sin_jugadores()  # Función que pide REB_O, REB_D, PERD, FC, etc.

    stats_equipo_completas = añadir_nombre_y_dorsal_stats("Equipo", "", stats_equipo)

    # 5. ORDENAR JUGADORES POR DORSAL ASCENDENTE
    stats_jugadores_completas = dict(
    sorted(
        equipo.items(),
        key=lambda x: float('inf') if x[1]['DORSAL'] == "" else int(x[1]['DORSAL'])
    )
)


    # 6. DEVOLVER DICCIONARIO COMPLETO
    return {
        "INFO_PARTIDO": {
            "EQUIPO_YO": nombre_equipo,
            "EQUIPO_RIVAL": nombre_rival,
            "COMPETICION": competicion,
            "TEMPORADA": temporada,
            "FECHA": fecha
        },
        "STATS_JUGADORES": stats_jugadores_completas,
        "STATS_EQUIPO": stats_equipo_completas
    }
```

## **FUNCIÓN 8: CALCULAR ESTADÍSTICA ACUMULADA DEL EQUIPO EN EL PARTIDO**



```python
def agregar_totales_equipo(jornada):
    """
    Calcula los totales del equipo a partir de:
    - stats_jugadores: diccionario con estadísticas de jugadores.
    - stats_equipo: diccionario con estadísticas no asignadas a jugadores.
    - totales: diccionario con estadísticas acumuladas del equipo.

    Añadimos stats_totales al diccionario global del partido (jornada).

    RESULTADO FUNCIÓN -> Devuelve jornada actualizada con totales.
    """

    stats_jugadores = jornada["STATS_JUGADORES"]
    stats_equipo = jornada ["STATS_EQUIPO"]

    # Inicializamos todos los campos a 0
    totales = {
        "NOMBRE": "Totales",
        "DORSAL": "",
        "PTS": 0,
        "TL_anot": 0, "TL_int": 0,
        "2P_anot": 0, "2P_int": 0,
        "3P_anot": 0, "3P_int": 0,
        "REB_O": 0, "REB_D": 0, "REB_TOT": 0,
        "AST": 0, "PERD": 0, "REC": 0,
        "TAP_F": 0, "TAP_C": 0,
        "MAT": 0,
        "FC": 0, "FR": 0,
        "PLUS_MINUS": 0,
        "VAL": 0,
        "MIN": 0
    }

    # Sumar estadísticas de JUGADORES.
    for j in stats_jugadores.values():
        for k in totales:
            if k in j and k not in ["NOMBRE", "DORSAL"]:
                totales[k] += j[k]

    # Sumar estadísticas de la fila EQUIPO.
    for k in totales:
        if k in stats_equipo and k not in ["NOMBRE", "DORSAL"]:
            totales[k] += stats_equipo[k]

    # Minutos MM:SS
    min_ent = int(totales["MIN"])
    seg = int(round((totales["MIN"] - min_ent) * 60))
    totales["MIN_MMSS"] = f"{min_ent}' {seg}''"

    # Añadimos los totales al diccionario global del partido
    jornada["STATS_TOTALES"] = totales
    return jornada
```

```python
def imprimir_info_partido(info):
    """
    Muestra por pantalla la información general del partido.
    """

    fecha_str = info["FECHA"].strftime("%d/%m/%Y")

    ancho_etiqueta = 16  # Ancho fijo para todas las etiquetas.

    print("\n" + "=" * 80)
    print(f"{'PARTIDO:':<{ancho_etiqueta}} {info['EQUIPO_YO']}  vs  {info['EQUIPO_RIVAL']}")
    print()
    print(f"{'COMPETICIÓN:':<{ancho_etiqueta}} {info['COMPETICION']}")
    print(f"{'TEMPORADA:':<{ancho_etiqueta}} {info['TEMPORADA']}")
    print(f"{'FECHA:':<{ancho_etiqueta}} {fecha_str}")
    print("=" * 80 + "\n")
```

## **FUNCIÓN 10: IMPRIMIR CADA FILA DEL BOXSCORE**


```python
# ---------------------- FUNCIÓN PARA IMPRIMIR CADA FILA DEL BOXSCORE ----------------------

def imprimir_fila(jugador, col_widths):
    """
    Imprime una fila del boxscore con formato ACB.
    """

    # Anchos de las columnas -> Para espaciar correctamente cada estadística.
    (col_dorsal, col_name, col_min, col_pts,
     col_shoot, col_pct, col_reb,
     col_other, col_mat, col_pm, col_val) = col_widths

    # Dorsal
    dorsal = str(jugador.get("DORSAL", "")).rjust(col_dorsal) #r.just -> Alineado a la derecha.

    # Si un nombre es más largo que el ancho prefijado para esa columna -> Recortamos el nombre. Para no descuadrar la tabla.
    nombre = jugador.get("NOMBRE", "")
    if len(nombre) > col_name:
        nombre = nombre[:col_name-3] + "."

    # Porcentajes de tiro -> Expresamos como enteros, para no descuadrar la tabla.
    pct_tl, pct_2p, pct_3p = porcentajes_tiro(jugador)
    pct_tl = int(round(pct_tl))
    pct_2p = int(round(pct_2p))
    pct_3p = int(round(pct_3p))

    # Tiros: anotados/intentados.
    t2 = f"{jugador.get('2P_anot',0)}/{jugador.get('2P_int',0)}"
    t3 = f"{jugador.get('3P_anot',0)}/{jugador.get('3P_int',0)}"
    tl = f"{jugador.get('TL_anot',0)}/{jugador.get('TL_int',0)}"

    # Imprimir fila.
    print(
        f"{dorsal:>{col_dorsal}} "
        f"{nombre:<{col_name}} "
        f"{jugador.get('MIN_MMSS',''):>{col_min}} "
        f"{jugador.get('PTS',0):>{col_pts}}   "
        f"{t2:>{col_shoot}} {pct_2p:>3}%    "
        f"{t3:>{col_shoot}} {pct_3p:>3}%    "
        f"{tl:>{col_shoot}} {pct_tl:>3}%     "
        f"{jugador.get('REB_O',0):>{col_reb}}  {jugador.get('REB_D',0):>{col_reb}}  {jugador.get('REB_TOT',0):>{col_reb}}    "
        f"{jugador.get('AST',0):>{col_other}}  {jugador.get('PERD',0):>{col_other}}  {jugador.get('REC',0):>{col_other}}    "
        f"{jugador.get('TAP_F',0):>{col_other}}  {jugador.get('TAP_C',0):>{col_other}}  {jugador.get('MAT',0):>{col_mat}}    "
        f"{jugador.get('FC',0):>{col_other}}  {jugador.get('FR',0):>{col_other}}    "
        f"{jugador.get('PLUS_MINUS',0):>{col_pm}}   "
        f"{jugador.get('VAL',0):>{col_pm}}"
    )
```

## **CALCULADORA 3: BOXSCORE DE UN PARTIDO**
## **CALCULADORA FINAL**

Modelo Boxscore: Liga Endesa (2025-2026).

Ejemplo:

https://live.acb.com/es/partidos/valencia-basket-vs-baxi-manresa-104584/estadisticas



```python
# ---------------------- CALCULADORA FINAL: MOSTRAR BOXSCORE COMPLETO ----------------------
# Boxscore estilo ACB (2025-2026). Tiene el formato siguiente:
# DORSAL  JUGADOR  | MIN | PTS | T2 | T2% | T3 | T3% | TL | TL% | RO | RD | RT | AS | PÉR | REC | TF | TR | MAT | FC | FR | +/- | Val


def imprimir_boxscore(jornada):
    """
    Muestra toda la información del partido + boxscore completo.
    """

    # 1. INFO DEL PARTIDO
    imprimir_info_partido(jornada["INFO_PARTIDO"])

    # 2. ANCHO DE COLUMNAS AJUSTADOS
    col_widths = (
        2,   # Dorsal
        22,  # Nombre
        9,   # MIN
        5,   # PTS
        5,   # Tiros (A/I)
        3,   # % Tiros
        3,   # Rebotes (RO/RD/RT)
        3,   # Otras stats (AST, PER, REC)
        4,   # MAT
        5,   # +/-
        6    # Val
    )

    # Línea de puntos "-" separadora. Ajusto manualmente el ancho, probando.
    linea = "-" * 180
    linea_bis = "=" * 180

    # 3. CABECERA DEL BOXSCORE -> Etiquetas de "JUGADOR", "PTS", etc.
    print("\n" + linea_bis)
    print(
        f"{'':>2} {'JUGADOR':<22} "
        f"{'MIN':>9} {'PTS':>5}   "
        f"{'T2':>4} {'T2%':>5}    "
        f"{'T3':>4} {'T3%':>5}    "
        f"{'TL':>4} {'TL%':>5}     "
        f"{'RO':>3}  {'RD':>3}  {'RT':>3}    "
        f"{'AS':>3}  {'PER':>3}  {'REC':>3}    "
        f"{'TF':>3}  {'TC':>3}  {'MAT':>4}    "
        f"{'FC':>3}  {'FR':>3}    "
        f"{'+/-':>5}   {'Val':>6}"
    )
    print(linea)

    # 4. JUGADORES (ordenados por dorsal).
    jugadores_ordenados = sorted(
        jornada["STATS_JUGADORES"].values(),
        key=lambda x: int(x.get("DORSAL", 0))
    )

    # IMPRIMIR UNA FILA POR CADA JUGADOR.
    for jugador in jugadores_ordenados:
        imprimir_fila(jugador, col_widths)

    print(linea)

    # 5. IMPRIMIR FILA DE EQUIPO.
    imprimir_fila(jornada["STATS_EQUIPO"], col_widths)
    print(linea)

    # 6. IMPRIMIR FILA DE TOTALES (ACUMULADO).
    imprimir_fila(jornada["STATS_TOTALES"], col_widths)
    print(linea_bis)
```

## **FUNCIÓN 11: PEDIR ESTADÍSTICAS DEL EQUIPO RIVAL**

```python
#Usamos la misma estructura que para pedir los datos de nuestros jugadores/equipo.

def agregar_stats_rival(jornada):
    print("\nIntroduce los datos del EQUIPO RIVAL:")

    # 1. PEDIR DATOS DEL EQUIPO RIVAL.
    while True:
        try:

            # MINUTOS JUGADOS -> Los mismos que mi equipo.

            # TIROS (anotados -> intentados).
            tl_anot = int(input("TL anotados: "))
            tl_int = int(input("TL intentados: "))
            t2_anot = int(input("2P anotados: "))
            t2_int = int(input("2P intentados: "))
            t3_anot = int(input("3P anotados: "))
            t3_int = int(input("3P intentados: "))

            # REBOTES.
            reb_o = int(input("Rebotes ofensivos: "))
            reb_d = int(input("Rebotes defensivos: "))

            # ASISTENCIAS, PERIDAS, RECUPERACIONES (i.e. robos).
            ast = int(input("Asistencias: "))
            perd = int(input("Pérdidas: "))
            rec = int(input("Recuperaciones: "))

            # TAPONES.
            tap_f = jornada["STATS_TOTALES"]["TAP_C"] #Los tapones a favor del rival son los tapones en contra de mi equipo, y al revés:
            tap_c = jornada["STATS_TOTALES"]["TAP_F"]

            # MATES.
            mates = int(input("Mates realizados: "))

            # FALTAS -> Igual que los tapones.
            fc = jornada["STATS_TOTALES"]["FR"] # Las faltas cometidas por el rival son las faltas recibidas por mi equipo. Y viceversa:
            fr = jornada["STATS_TOTALES"]["FC"] # Las faltas recibidas por el rival son las faltas cometidas por mi equipo.


            # CÁLCULO DE TOTALES: Puntos, rebotes & valoración ACB.
            pts = tl_anot + t2_anot*2 + t3_anot*3
            reb_tot = reb_o + reb_d

            tiros_fallados = (tl_int - tl_anot) + (t2_int - t2_anot) + (t3_int - t3_anot)
            valoracion = pts + reb_tot + ast + rec + tap_f + fr - ( tiros_fallados + perd + tap_c + fc)

            # -------------------------------------------------
            # 2 VALIDACIONES.
            # -------------------------------------------------

            # 2.1 - NO VALORES NEGATIVOS.
            if min(tl_anot, tl_int, t2_anot, t2_int,
                   t3_anot, t3_int, reb_o, reb_d, ast, perd, rec,
                   tap_f, tap_c, mates, fc, fr) < 0:
                print("ERROR. No se permiten valores negativos.")
                continue

            # 2.2 - TIROS COHERENTES.
            if tl_anot > tl_int or t2_anot > t2_int or t3_anot > t3_int:
                print("ERROR. Los tiros anotados no pueden superar a los intentados.")
                continue

            # 2.3 - TAPONES RECIBIDOS COHERENTES CON TIROS DE CAMPO INTENTADOS.
            tiros_campo_int = t2_int + t3_int
            if tap_c > tiros_campo_int:
                print("ERROR. No se pueden recibir más tapones que tiros intentados.")
                continue

            break

        # 2.4 - DATOS DE TIPO 'int'.
        except ValueError:
            print("ERROR. Introduce valores numéricos válidos.\n")

    # 3. RESULTADO FUNCIÓN -> ESTADÍSTICAS EN DICCIONARIO. Mismo formato que para nuestro equipo.
    # Pido todas las estadísticas, aunque quizás luego no use todas, para mantener estructura de los datos.
    # Si luego necesito un dato particular, ya lo tengo, no necesito pedir nada más.

    totales_rival = {
        "PTS": pts,
        "TL_anot": tl_anot, "TL_int": tl_int,
        "2P_anot": t2_anot, "2P_int": t2_int,
        "3P_anot": t3_anot, "3P_int": t3_int,
        "REB_O": reb_o, "REB_D": reb_d, "REB_TOT": reb_tot,
        "AST": ast, "PERD": perd, "REC": rec,
        "TAP_F": tap_f, "TAP_C": tap_c,
        "MAT":mates,
        "FC": fc, "FR": fr,
        "PLUS_MINUS": 0,
        "VAL": valoracion,
        "MIN": jornada["STATS_TOTALES"]["MIN"],
        "MIN_MMSS": jornada["STATS_TOTALES"]["MIN_MMSS"]
    }

    # Añadimos los totales al diccionario global del partido.
    jornada["STATS_TOTALES_RIVAL"] = totales_rival
    return jornada
```

## **FUNCIÓN 12: ESTADÍSTICA AVANZADA - EQUIPO**

```python
def calcular_stats_avanzadas_equipo(tot, rival):

    FG_anot = tot["2P_anot"] + tot["3P_anot"]
    FG_int = tot["2P_int"] + tot["3P_int"]

    # POSESIONES -> No tenemos un dato exacto de posesiones de cada equipo.
    # Recurrimos a la fórmula aproximada de las posesiones. Para que coincidan OER_equipo con DER_rival / DER_equipo con OER_Rival,
    # calculamos la media de las posesiones

    posesiones_mi_equipo = FG_int + 0.44 * tot["TL_int"] - tot["REB_O"] + tot["PERD"]
    posesiones_rival = rival["2P_int"] + rival["3P_int"] + 0.44 * rival["TL_int"] - rival["REB_O"] + rival["PERD"]

    posesiones = (posesiones_mi_equipo + posesiones_rival)/2


    # OER -> Rendimiento Ofensivo.
    oer = (tot["PTS"] / posesiones)*100 if posesiones > 0 else 0

    # DER -> Rendimiento Defensivo.
    der = (rival["PTS"] / posesiones)*100  if posesiones > 0 else 0

    # NET_RATING -> Rendimiento Neto: Ofensivo-Defensivo.


    # FOUR FACTORS.
    efg = 0
    if FG_int > 0:
        efg = ((FG_anot + 0.5 * tot["3P_anot"]) / FG_int) * 100

    tov_pct = 0
    if posesiones > 0:
        tov_pct = (tot["PERD"] / posesiones) * 100

    ft_rate = 0
    if FG_int > 0:
        ft_rate = (tot["TL_anot"] / FG_int)

    oreb_pct = 0
    if (tot["REB_O"] + rival["REB_D"]) > 0:
        oreb_pct = (tot["REB_O"] / (tot["REB_O"] + rival["REB_D"])) * 100

    # PORCENTAJE REBOTE DEFENSIVO.
    dreb_pct = 0
    if (tot["REB_D"] + rival["REB_O"]) > 0:
        dreb_pct = (tot["REB_D"] / (tot["REB_D"] + rival["REB_O"])) * 100

    # TRUE SHOOTING.
    ts = 0
    if (FG_int + 0.44 * tot["TL_int"]) > 0:
        ts = (tot["PTS"]*0.5 / (FG_int + 0.44 * tot["TL_int"])) * 100

    return {
        "POSESIONES": posesiones,
        "OER": oer,
        "DER": der,
        "NET": oer - der,
        "eFG%": efg,
        "TOV%": tov_pct,
        "FT_RATE": ft_rate,
        "OREB%": oreb_pct,
        "DREB%": dreb_pct,
        "TS%": ts
    }
```

## **FUNCIÓN 13: ESTADÍSTICA AVANZADA - JUGADORES**


```python
def calcular_stats_avanzadas_jugador(j, tot):

    #j   -> Estadísticas del jugador.
    #tot -> Estadísticas totales del equipo.

    # Tiros intentados
    fg_int_j = j["2P_int"] + j["3P_int"]
    fg_anot_j = j["2P_anot"] + j["3P_anot"]

    # OER
    oer = 0
    posesiones_j = fg_int_j + 0.44 * j["TL_int"] - j["REB_O"] + j["PERD"]
    if posesiones_j > 0:
        oer = (j["PTS"] / posesiones_j) * 100

    # eFG%
    efg = 0
    if fg_int_j > 0:
        efg = ((fg_anot_j + 0.5 * j["3P_anot"]) / fg_int_j) * 100

    # TS%
    ts = 0
    ts_denom = fg_int_j + 0.44 * j["TL_int"]
    if ts_denom > 0:
        ts = (j["PTS"] * 0.5 / (ts_denom))*100


    # AST%
    ast_denom = ( j["MIN"] / (tot["MIN"]/5))*(tot["2P_anot"] + tot["3P_anot"]) - fg_anot_j
    ast_pct = 0
    if ast_denom > 0:
        ast_pct = (j["AST"] / ast_denom) * 100

    # AST/TO
    ast_to = 0
    if j["PERD"] > 0:
        ast_to = j["AST"] / j["PERD"]


    # USG%
    usg_denom = (tot["2P_int"] + tot["3P_int"] + 0.44 * tot["TL_int"] + tot["PERD"]) * (j["MIN"] / (tot["MIN"]/5))
    usg = 0
    if usg_denom > 0:
        usg = ( (fg_int_j + 0.44 * j["TL_int"] + j["PERD"]) / usg_denom ) * 100

    return {
        "OER": oer,
        "eFG%": efg,
        "TS%": ts,
        "AST%": ast_pct,
        "AST/TO": ast_to,
        "USG%": usg
    }
```

## **CALCULADORA 4: ESTADÍSTICA AVANZADA**
## **CALCULADORA FINAL**

```python
def calculadora_estadistica_avanzada(jornada):

    tot = jornada["STATS_TOTALES"]
    rival = jornada["STATS_TOTALES_RIVAL"]

    tot["AVANZADAS"] = calcular_stats_avanzadas_equipo(tot, rival)
    rival["AVANZADAS"] = calcular_stats_avanzadas_equipo(rival, tot)


    for j in jornada["STATS_JUGADORES"].values():
        j["AVANZADAS"] = calcular_stats_avanzadas_jugador(j, tot)
```

## **FUNCIÓN 14: ACUMULAR ESTADÍSTICAS NUMÉRICAS**

```python
def acumular_stats(destino, origen):
    """
    Suma las estadísticas numéricas de 'origen' en 'destino'.
    Las claves no numéricas se copian solo una vez.
    """

    for clave, valor in origen.items():

        if isinstance(valor, (int, float)):
            destino[clave] = destino.get(clave, 0) + valor
        else:
            destino.setdefault(clave, valor)
```

## **FUNCIÓN 15: CONVERTIR MINUTOS**

```python
def minutos_a_mmss(minutos):
    """
    Convierte minutos en formato float a string MM' SS''
    Ej: 40.5 -> 40' 30''
    """
    total_segundos = int(round(minutos * 60))
    mm = total_segundos // 60
    ss = total_segundos % 60
    return f"{mm}' {ss:02d}''"
```

## **CALCULADORA 5**
## **CALCULADORA FINAL -> ESTADÍSTICAS ACUMULADAS TEMPORADA**

```python
def calculadora_temporada_totales(jornadas):
    """
    Calcula estadísticas acumuladas de una temporada completa.
    """

    if not jornadas:
        raise ValueError("La lista de jornadas está vacía.")

    # --------------------------------
    # 1. REFERENCIAS DE CONTROL. -> LOS PARTIDOS DEBEN SER DEL MISMO EQUIPO Y DE LA MISMA COMPETICIÓN.
    # --------------------------------
    equipo_ref = jornadas[0]["INFO_PARTIDO"]["EQUIPO_YO"]
    competicion_ref = jornadas[0]["INFO_PARTIDO"]["COMPETICION"]
    temporada_ref = jornadas[0]["INFO_PARTIDO"]["TEMPORADA"]
    fecha_ref = jornadas[0]["INFO_PARTIDO"]["FECHA"]


    # --------------------------------
    # 2. ESTRUCTURA DE LAS STATS DE LA TEMPORADA -> LA MISMA QUE LA DE CADA JORNADA, PARA PODER REUTILIZAR LAS FUNCIONES YA PROGRAMADAS.
    # --------------------------------
    temporada = {
        "INFO_PARTIDO": {
            "EQUIPO_YO": equipo_ref,
            "EQUIPO_RIVAL": "Rivales",
            "COMPETICION": competicion_ref,
            'TEMPORADA': temporada_ref,
            'FECHA': fecha_ref
        },
        "STATS_JUGADORES": {},
        "STATS_EQUIPO": {},
        "STATS_TOTALES": {},
        "STATS_TOTALES_RIVAL": {}
    }

    # --------------------------------
    # 3. RECORRER JORNADAS.
    # --------------------------------
    for jornada in jornadas:

        # -------- VALIDACIONES --------
        if jornada["INFO_PARTIDO"]["EQUIPO_YO"] != equipo_ref:
            raise ValueError("Estas calculando estadísticas de equipos diferentes. Comprueba los datos.")

        if jornada["INFO_PARTIDO"]["COMPETICION"] != competicion_ref:
            raise ValueError("Estas calculando estadísticas de competiciones distintas. Comprueba los datos.")

        # -------- EQUIPO -> Fila boxscore --------
        acumular_stats(
            temporada["STATS_EQUIPO"],
            jornada["STATS_EQUIPO"]
        )

        # -------- EQUIPO -> Acumuladas --------
        acumular_stats(
            temporada["STATS_TOTALES"],
            jornada["STATS_TOTALES"]
        )

        # -------- RIVAL --------
        acumular_stats(
            temporada["STATS_TOTALES_RIVAL"],
            jornada["STATS_TOTALES_RIVAL"]
        )

        # -------- JUGADORES --------
        # Se identifican por nombre. Si un jugador ha jugado dos partidos con dorsales diferentes (Ej: Kobe Bryant con el '8' y el '24')
        # hay que contabilizar sus estadísticas totales. Es el mismo jugador!

        for jugador in jornada["STATS_JUGADORES"].values():

            nombre = jugador["NOMBRE"]

            # Primera vez que aparece el jugador.
            if nombre not in temporada["STATS_JUGADORES"]:
                nuevo = jugador.copy()
                nuevo["PJ"] = 1
                temporada["STATS_JUGADORES"][nombre] = nuevo

            # Ya existe → acumular stats
            else:
                acum = temporada["STATS_JUGADORES"][nombre]
                acumular_stats(acum, jugador)
                acum["PJ"] += 1

    # --------------------------------
    # 4. RECALCULAR MIN_MMSS
    # --------------------------------

    # Equipo
    if "MIN" in temporada["STATS_TOTALES"]:
        temporada["STATS_TOTALES"]["MIN_MMSS"] = minutos_a_mmss(
            temporada["STATS_TOTALES"]["MIN"]
        )

    # Rival
    if "MIN" in temporada["STATS_TOTALES_RIVAL"]:
        temporada["STATS_TOTALES_RIVAL"]["MIN_MMSS"] = minutos_a_mmss(
            temporada["STATS_TOTALES_RIVAL"]["MIN"]
        )

    # Jugadores
    for jugador in temporada["STATS_JUGADORES"].values():
        jugador["MIN_MMSS"] = minutos_a_mmss(jugador["MIN"])


    return temporada
```

```python
def calculadora_temporada_medias(temporada, num_partidos):
    """
    Calcula las medias por partido de una temporada,
    teniendo en cuenta:
    - PJ de cada jugador. para jugadores
    - Número total de partidos del equipo (para TOTALES)
    Devuelve un diccionario con el mismo formato que una jornada, para poder reutilizar funciones ya programadas.
    """

    medias = {}

    # --------------------------------
    # 1. INFO PARTIDO -> COPIAR DATOS.
    # --------------------------------
    medias["INFO_PARTIDO"] = temporada["INFO_PARTIDO"].copy()

    # --------------------------------
    # 2. JUGADORES (media por PJ).
    # --------------------------------
    medias["STATS_JUGADORES"] = {}

    for nombre, j in temporada["STATS_JUGADORES"].items():
        pj = j["PJ"]
        if pj == 0:
            continue

        jugador_media = {}

        for clave, valor in j.items():
            if clave in ["NOMBRE", "DORSAL", "PJ"]:
                jugador_media[clave] = valor
            elif clave == "MIN":
                jugador_media["MIN"] = valor / pj
            elif clave == "MIN_MMSS":
                continue
            else:
                if isinstance(valor, (int, float)):
                    jugador_media[clave] = valor / pj
                else:
                    jugador_media[clave] = valor

        jugador_media["MIN_MMSS"] = minutos_a_mmss(jugador_media["MIN"])
        medias["STATS_JUGADORES"][nombre] = jugador_media

    # --------------------------------
    # 3. EQUIPO (media por partido)
    # --------------------------------
    medias["STATS_EQUIPO"] = {}

    for clave, valor in temporada["STATS_EQUIPO"].items():
        if clave in ["NOMBRE", "DORSAL"]:
            medias["STATS_EQUIPO"][clave] = valor
        elif clave == "MIN":
            medias["STATS_EQUIPO"]["MIN"] = valor / num_partidos
        elif clave == "MIN_MMSS":
            continue
        else:
            if isinstance(valor, (int, float)):
                medias["STATS_EQUIPO"][clave] = valor / num_partidos
            else:
                medias["STATS_EQUIPO"][clave] = valor

    medias["STATS_EQUIPO"]["MIN_MMSS"] = minutos_a_mmss(
        medias["STATS_EQUIPO"]["MIN"]
    )

    # --------------------------------
    # 4. TOTALES (media por partido)
    # --------------------------------
    medias["STATS_TOTALES"] = {}

    for clave, valor in temporada["STATS_TOTALES"].items():
        if clave in ["NOMBRE", "DORSAL"]:
            medias["STATS_TOTALES"][clave] = valor
        elif clave == "MIN":
            medias["STATS_TOTALES"]["MIN"] = valor / num_partidos
        elif clave == "MIN_MMSS":
            continue
        else:
            if isinstance(valor, (int, float)):
                medias["STATS_TOTALES"][clave] = valor / num_partidos
            else:
                medias["STATS_TOTALES"][clave] = valor

    medias["STATS_TOTALES"]["MIN_MMSS"] = minutos_a_mmss(
        medias["STATS_TOTALES"]["MIN"]
    )

    return medias
```

```python
def imprimir_fila_medias(jugador, col_widths,numero_partidos):
    """
    Imprime una fila del boxscore con formato ACB.
    """

    # Anchos de las columnas -> Para espaciar correctamente cada estadística.
    (col_dorsal, col_name, col_pj, col_min, col_pts,
     col_shoot, col_pct, col_reb,
     col_other, col_mat, col_pm, col_val) = col_widths

    # Dorsal
    dorsal = str(jugador.get("DORSAL", "")).rjust(col_dorsal) #r.just -> Alineado a la derecha.

    # Si un nombre es más largo que el ancho prefijado para esa columna -> Recortamos el nombre. Para no descuadrar la tabla.
    nombre = jugador.get("NOMBRE", "")
    if len(nombre) > col_name:
        nombre = nombre[:col_name-3] + "."

    # Porcentajes de tiro -> Expresamos como enteros, para no descuadrar la tabla.
    pct_tl, pct_2p, pct_3p = porcentajes_tiro(jugador)
    pct_tl = round(pct_tl)
    pct_2p = round(pct_2p)
    pct_3p = round(pct_3p)

    # Tiros: anotados/intentados.
    t2 = f"{round(jugador.get('2P_anot',0))}/{round(jugador.get('2P_int',0))}"
    t3 = f"{round(jugador.get('3P_anot',0))}/{round(jugador.get('3P_int',0))}"
    tl = f"{round(jugador.get('TL_anot',0))}/{round(jugador.get('TL_int',0))}"

    # Imprimir fila.
    print(
        f"{dorsal:>{col_dorsal}} "
        f"{nombre:<{col_name}} "
        f"{jugador.get('PJ',numero_partidos):>{col_pj}}"
        f"{jugador.get('MIN_MMSS',''):>{col_min}} "
        f"{round(jugador.get('PTS',0),1):>{col_pts}}   "
        f"{t2:>{col_shoot}} {pct_2p:>3}%    "
        f"{t3:>{col_shoot}} {pct_3p:>3}%    "
        f"{tl:>{col_shoot}} {pct_tl:>3}%     "
        f"{round(jugador.get('REB_O',0),1):>{col_reb}}  {round(jugador.get('REB_D',0),1):>{col_reb}}  {round(jugador.get('REB_TOT',0),1):>{col_reb}}    "
        f"{round(jugador.get('AST',0),1):>{col_other}}  {round(jugador.get('PERD',0),1):>{col_other},}  {round(jugador.get('REC',0),1):>{col_other}}    "
        f"{round(jugador.get('TAP_F',0),1):>{col_other}}  {round(jugador.get('TAP_C',0),1):>{col_other}}  {round(jugador.get('MAT',0),1):>{col_mat}}    "
        f"{round(jugador.get('FC',0),1):>{col_other}}  {round(jugador.get('FR',0),1):>{col_other}}    "
        f"{round(jugador.get('PLUS_MINUS',0),1):>{col_pm}}   "
        f"{round(jugador.get('VAL',0),1):>{col_pm}}"
    )
```

```python
def imprimir_boxscore_medias(jornada,numero_partidos):
    """
    Muestra toda la información del partido + boxscore completo.
    """

    # 1. INFO DEL PARTIDO
    imprimir_info_partido(jornada["INFO_PARTIDO"])

    # 2. ANCHO DE COLUMNAS AJUSTADOS
    col_widths = (
        2,   # Dorsal
        22,  # Nombre
        3,   # PJ
        10,   # MIN
        5,   # PTS
        5,   # Tiros (A/I)
        3,   # % Tiros
        3,   # Rebotes (RO/RD/RT)
        3,   # Otras stats (AST, PER, REC)
        4,   # MAT
        5,   # +/-
        6    # Val
    )

    # Línea de puntos "-" separadora. Ajusto manualmente el ancho, probando.
    linea = "-" * 180
    linea_bis = "=" * 180

    # 3. CABECERA DEL BOXSCORE -> Etiquetas de "JUGADOR", "PTS", etc.
    print("\n" + linea_bis)
    print(
        f"{'':>2} {'JUGADOR':<22} "
        f"{'PJ':<3} "
        f"{'MIN':>9} {'PTS':>5}   "
        f"{'T2':>4} {'T2%':>5}    "
        f"{'T3':>4} {'T3%':>5}    "
        f"{'TL':>4} {'TL%':>5}     "
        f"{'RO':>3}  {'RD':>3}  {'RT':>3}    "
        f"{'AS':>3}  {'PER':>3}  {'REC':>3}    "
        f"{'TF':>3}  {'TC':>3}  {'MAT':>4}    "
        f"{'FC':>3}  {'FR':>3}    "
        f"{'+/-':>5}   {'Val':>6}"
    )
    print(linea)

    # 4. JUGADORES (ordenados por dorsal).
    jugadores_ordenados = sorted(
        jornada["STATS_JUGADORES"].values(),
        key=lambda x: int(x.get("DORSAL", 0))
    )

    # IMPRIMIR UNA FILA POR CADA JUGADOR.
    for jugador in jugadores_ordenados:
        imprimir_fila_medias(jugador, col_widths,numero_partidos)

    print(linea)

    # 5. IMPRIMIR FILA DE EQUIPO.
    imprimir_fila_medias(jornada["STATS_EQUIPO"], col_widths,numero_partidos)
    print(linea)

    # 6. IMPRIMIR FILA DE TOTALES (ACUMULADO).
    imprimir_fila_medias(jornada["STATS_TOTALES"], col_widths,numero_partidos)
    print(linea_bis)
```
